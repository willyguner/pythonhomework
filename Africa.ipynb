# 後勤最大利益分析


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQdsrNhtjOUpS1lhdofAjV_aSB2M4C2wxScMN8dISgSoP1Dg1FL_susncv0ieKXoFAqff_vn1G5Fhd1/pub?gid=2016644013&single=true&output=csv')
```

```python
df = df.fillna(0) #NaN值補 0
```

```python
df.head()
```

## 標準化 及 加權

此遊戲的零件3分鐘回1點，故須將零件*3 

道具部分，因為為隨機取得，且就算獲得一次後勤只能獲得其中一個，故不列入分析。

```python
df.insert(7,'零件(加權後)',df.零件*3)
df.insert(8,'總和',df.人口+df.彈藥+df.口糧+df['零件(加權後)'])
df['人口(每分鐘)']=df['人口']/df['時間(分鐘)']
df['彈藥(每分鐘)']=df['彈藥']/df['時間(分鐘)']
df['口糧(每分鐘)']=df['口糧']/df['時間(分鐘)']
df['零件(每分鐘)']=df['零件']/df['時間(分鐘)']
df.head()
```

# 實際需求

根據我玩這個遊戲三年的經驗，這遊戲最缺的就是零件，其次就是口糧，因此我們來看看如何才能讓這兩個資源收益最高。

```python
df=df.sort_values(by=['零件(每分鐘)','口糧(每分鐘)'],ascending=[False,False])
df
```

# 最大利益圖表

最後由於一次最多只能出4隊後勤，將前4項總和加總，並假設沒有任何時間被浪費掉，固可得圖表。

```python
people=0
ammo=0
ration=0
component=0


for i in range(4):
    people+=df['人口(每分鐘)'].values[i]*180
    ammo+=df['彈藥(每分鐘)'].values[i]*180
    ration+=df['口糧(每分鐘)'].values[i]*180
    component+=df['零件(每分鐘)'].values[i]*180
    print(df['章節'].values[i])

bar=plt.bar(['people','ammo','ration','component'],[people,ammo,ration,component])

def createLabels(data):
    for item in data:
        height = item.get_height()
        plt.text(
            item.get_x()+item.get_width()/2., 
            height*1.05, 
            '%d' % int(height),
            ha = "center",
            va = "bottom",
        )
        
createLabels(bar)
```

此為三個小時中可得的最大資源，以及章節
