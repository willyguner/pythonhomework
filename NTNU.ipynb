```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1=np.linspace(-(np.pi/2),5*(np.pi/2),100)
y1=np.sin(x1)
plt.plot(x1,y1)

plt.plot([9,17],[1,1])
plt.plot([13,13],[1,-1])

x3=np.linspace(11*(np.pi/2),17*(np.pi/2),100)
y3=np.sin(x3)
plt.plot(x3,y3)

x4=np.linspace(30,40)
y4=8*(((x4-35)**2)/100)-1
plt.plot(x4,y4)
```
